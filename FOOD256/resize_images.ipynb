```python
import os, csv
from PIL import Image
```

```python
folders = [line for line in os.listdir('data') if '.' not in line]
for prog, folder in enumerate(folders):
    bboxes = {}
    with open('./data/' + folder + '/bb_info.txt', 'rb') as bbox_data:
        reader = csv.reader(bbox_data)
        bbox_rows = [row for row in reader][1:]
        for ii, row in enumerate(bbox_rows):
            bound = [int(elem) for elem in row[0].split()]
            bboxes[str(bound[0])] = bound[1:]
    print '1 - %s: loaded %d bounding boxes' % (folder, len(bboxes.keys()))
    imageFiles = [line for line in os.listdir('data/' + folder) if '.jpg' in line and '-res' not in line]
    
    print '2 - %s: loaded %d images' % (folder, len(imageFiles))
    assert len(imageFiles) == len(bboxes.keys())
    for fileName in imageFiles:
        imageID = fileName.split('.')[0]
        fullName = './data/%s/%s' % (folder, fileName)
#         print fullName
        im = Image.open(fullName)
        cropped = im.crop(tuple(bboxes[imageID]))

        # http://matthiaseisen.com/pp/patterns/p0202/
#         longer_side = max(cropped.size)
#         horizontal_padding = (longer_side - cropped.size[0]) / 2
#         vertical_padding = (longer_side - cropped.size[1]) / 2
#         padded = cropped.crop(
#             (
#                 -horizontal_padding,
#                 -vertical_padding,
#                 cropped.size[0] + horizontal_padding,
#                 cropped.size[1] + vertical_padding
#             )
#         )
#         print padded.size
        resized = cropped.resize((299, 299), Image.ANTIALIAS)
        resized.save('./data/%s/%s-filled.jpg' % (folder, imageID))
    print '3 - %s: cropped %d images' % (folder, len(imageFiles))
    print '%d/%d' % (prog + 1, len(folders))
    
```
